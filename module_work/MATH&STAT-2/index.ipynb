```python

import numpy as np # для работы с массивами
import pandas as pd # для работы DataFrame
import seaborn as sns # библиотека для визуализации статистических данных
import matplotlib.pyplot as plt # для построения графиков
```

```python
student = pd.read_csv('Admission_Predict_Ver1.1.csv')

del student['Serial No.']

corr = student.corr()

print(corr)
```

```python
plt.figure(figsize=(16,6)) # размер графика
sns.heatmap(data=corr, annot=True)
```

```python
Research_yes = student[student['Research'] == 1]
Research_no = student[student['Research'] == 0]
```

```python
plt.figure(figsize=(16,6)) # размер графика
sns.heatmap(data=Research_yes.corr(), annot=True)
```

```python
plt.figure(figsize=(16,6)) # размер графика
sns.heatmap(data=Research_no.corr(), annot=True)
```
